```python
##Initialize an empty list to store the items whose datatype is object
c_list=[]

##Initialize an empty list to store the items whose datatype is float and int
n_list=[]

##Initialize an empty list to store the items whose datatype is datetime
d_list=[]

for col_name in df.columns.tolist():
  if str(df[col_name].dtype) == "object":
    c_list.append(col_name)
  elif str(df[col_name].dtype) in ['float64', 'int64']:
    n_list.append(col_name)
  else:
    d_list.append(col_name)
```

```python
#pd.factorize - used for obtaining a numeric representation of an array when all that matters is identifying distinct values .
for col_name in c_list:
  df[col_name],fact=pd.factorize(df[col_name])
  print(fact)
```

**DIMENSIONALITY REDUCTION**

```python
drop_list=['Sr.No','Date of Birth', 'Exit Date', 'Exit Type','Date of Joining','Date of Retirement']
dropped_data = df.drop(drop_list,axis=1)
```

```python
dropped_data.fillna(-1,inplace=True)
```

```python
dropped_data.info()
```

```python
X=dropped_data.drop(['Curr Status'],axis=1)
Y=dropped_data['Curr Status']
```

```python
X
```

```python
Y
```

```python
##XGBoost Model
model = XGBClassifier()
sum(cross_val_score(model,X,Y,cv=10,scoring='accuracy'))/10
```

```python
Y.value_counts()
```

```python

# split data into train and test sets
seed = 7
test_size = 0.20
train_x, test_x, train_y, test_y = train_test_split(X,Y,test_size=test_size,random_state=seed)
```

```python
#Fit the model on trainining data
model.fit(train_x,train_y)

#Make predictions for test data
predict = model.predict(test_x)

#Evaluate predictions 
print(accuracy_score(test_y,predict))
```

```python
rpred=model.predict(X)
print(accuracy_score(Y,rpred))
```
